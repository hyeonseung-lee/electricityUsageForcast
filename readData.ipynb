### 그렇다면 ESS 설비를 갖춘 건물과 갖추지 못한 건물의 시간별 전력 소모량은 어떻게 될까?

#### 일단 시간 데이터를 전처리 해주자. 
날짜와 시간을 분리해서 저장한다

```python
usage_df
usage_df['일시'] = pd.to_datetime(usage_df['일시'], format="%Y%m%d %H")
# 날짜 컬럼 추가
usage_df['날짜'] = usage_df['일시'].dt.date

# 시간 컬럼 추가
usage_df['시간'] = usage_df['일시'].dt.time
usage_df
```

설비를 갖춘 건물의 경우

```python
import matplotlib.pyplot as plt
```

```python
building_with_ESS = isThereESS_df['건물번호'].reset_index(drop=True)
building_with_ESS
```

```python
BWS1 = building_with_ESS[0]
BWS1
bws1_usage = usage_df[usage_df['건물번호'] == BWS1]
bws1_usage
```

전체 그래프

```python
# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(bws1_usage['일시'], bws1_usage['전력소비량(kWh)'], marker='o', linestyle='-')
plt.xlabel('일시')
plt.ylabel('전력소비량(kWh)')
plt.title('일시별 전력소비량(kWh)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
